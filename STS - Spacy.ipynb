```python
import spacy
from sklearn.metrics.pairwise import cosine_similarity
import pandas as pd
```

```python
nlp_large = spacy.load("es_core_news_lg")
```

```python
sentences = [
    'En que puedo ayudar',
    'Como puedo ayudarlo',
    'Puedo ayudarle',
    'Como puedo asistirlo',
    'Como puedo robarlo',
    'Que queres',
    'No molestes',
    'Que necesita resolver',
    'Mi abuela tiene un biombo'
]

print('\n'.join(sentences))
```

```python
vectors_large = [nlp_large(sentence).vector for sentence in sentences]

similarities_large = cosine_similarity(vectors_large)
a_lg=pd.DataFrame(similarities_large,
            index=sentences,
            columns=sentences) 
a_lg.corr().style.background_gradient(cmap="coolwarm")
```
